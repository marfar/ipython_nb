\begin{equation}
\mu_U \frac{R_{on}}{D} U(t) \theta(\omega)\theta(D-\omega) =
\frac{d\omega}{dt}\Big[ R_{on}\frac{\omega(t)}{D} + R_{off} \Big(1-\frac{\omega(t)}{D}\Big)  \Big]
\end{equation}
Зведемо рівняння до безрозмірного вигляду, ввівши безрозмірні величини $X(t)=\frac{\omega(t)}{D}$, $R_i = \frac{R_{off}}{R_{on}}$, $u = \frac{\mu_U U(t)}{D^2}$.
Тоді рівнянн набуде наступного вигляду:
\begin{equation}
u(t) \theta(X)\theta(1-X) =
\frac{dX}{dt}\Big[ X(t) + R_i \Big(1-X(t)\Big)  \Big]
\end{equation}


# Обчислювальне завдання №1

```python
#!/bin/env/python3
from pylab import *
%matplotlib inline
from IPython.display import display, Math, Latex

def align_yaxis(ax1, v1, ax2, v2):
	"""adjust ax2 ylimit so that v2 in ax2 is aligned to v1 in ax1"""
	_, y1 = ax1.transData.transform((0, v1))
	_, y2 = ax2.transData.transform((0, v2))
	inv = ax2.transData.inverted()
	_, dy = inv.transform((0, 0)) - inv.transform((0, y1-y2))
	miny, maxy = ax2.get_ylim()
	ax2.set_ylim(miny+dy, maxy+dy)

def runge_kutta4(t, X, f, h):
	k0 = h * f(t, X)
	k1 = h * f(t + h/2, X + k0/2)
	k2 = h * f(t + h/2, X + k1/2)
	k3 = h * f(t + h, X + k2)
	X = X + 1/6.*( k0 + 2*k1 + 2*k2 + k3)
	return X

def memristor(t, u, mu_U=10**-6, x0=0, D=10, r_on=0.1,
			  r_off=16, v0=100, F=lambda X: 1):
	#Memristor resistance
	#print(t)
	
	#h = (t[1] - t[0])/100
	f = lambda t, X: u(t) * F(X) /	( X + R_i * (1.- X) )

	R_i = r_off/r_on
	U0_max = D*v0*r_off/(4*mu_U/D*(1-r_on/r_off)*r_on)
	
	display(Latex(r'$U_0^{max}~ = ~'+str(round(U0_max,2))+' ~V$'))
	
	X = zeros(len(t)) + x0
	
	for i in range(1,len(t)):
		#print(t[i],t[i-1],w[-1])
		#if w[-1] == 0 : w[-1]=rand(1)*D/1000
		eps = 10**-15
		h = (t[i]-t[i-1])
		X[i] = runge_kutta4(t[i-1], X[i-1], f, h)
		if X[i] >= 1.: X[i] = (1-rand(1)*eps)
		elif X[i] <= 0.: X[i] = eps*rand(1)
		else: pass
		#if X[i]==X[i-1]: X[i]+=((X[i]<=0) - (X[i]>=1))*rand(1)*eps
	#X = X *(X>0)*(X<1) + X/X*(X>1)
	U = u(t) / mu_U * D**2
	r = (r_on*X + r_off*(1-X))

	I = u(t)/r/1000 # A

	q = zeros(len(t))
	qf = lambda i, q: I[i]
	for i in range(1,len(t)):
		h = t[i]-t[i-1]
		q[i] = runge_kutta4(i, q[i-1], qf, h)

	F = zeros(len(t))
	Ff = lambda i, F: U[i]
	for i in range(1,len(t)):
		h = t[i]-t[i-1]
		F[i] = runge_kutta4(i, F[i-1], Ff, h)

	
	import matplotlib.gridspec as gridspec
	
	gs = gridspec.GridSpec(6, 2)
	
	figure(figsize=(15,10))
	ax1=subplot(gs[0,0])
	ax1.plot(t,U,'g')
	ax2 = ax1.twinx()
	ax2.plot(t,I,'r',t,[0]*len(t),'k--')
	ax1.set_ylabel('U, V', color='green')
	ax2.set_ylabel('I, A', color='red')

	for tl in ax1.get_yticklabels():
		tl.set_color('green')
	for tl in ax2.get_yticklabels():
		tl.set_color('red')
	ax1.grid(1,color='g')
	ax2.grid(1,color='r')
	align_yaxis(ax1,0,ax2,0)

	subplot(gs[1,0])
	plot(t,r)
	ylabel('R, kOhm')
	grid(1)

	subplot(gs[2,0])
	ylabel('w/D')
	plot(t,X)
	xlabel('t, sec')
	grid(1); gs.update( wspace=.7, hspace=0.8)

	subplot(gs[3:6,0])
	plot(U, I)
	ylabel('I, A')
	xlabel('U, V')
	grid(1)

	subplot(gs[0:3,1])
	plot(U, r)
	xlabel('U, V')
	ylabel('R, kOhm')
	grid(1)

	subplot(gs[3:6,1])
	plot(F, q)
	xlabel(r'$\Phi$')
	ylabel('q')
	grid(1)
	#show(0)		
	
	return X, U, I, r
```

#### З вхідними даними з статті:

```python
	
##########################################################################################
##########################################################################################
W0 = 1. 	# Hz
mu_U = 10**4 	# nm**2/s/V
D = 60. 	# nm

r_on = 0.1 	#kOhm
r_off = 16 	#kOhm
R_i = r_off/r_on
r0 = 4 	#R(t=0), kOhm
U0 = -1.5 	# V
x0 = (r0 - r_off)/(r_on - r_off) 	# start X 

t = linspace(0,2*pi/W0*3,2000)
u = lambda t_: U0*sin(W0*t_) * mu_U / D**2


out = memristor(t=t,u=u,v0=W0, x0=x0, r_on=r_on, r_off=r_off, D=D, mu_U=mu_U)
#X = X*(X>-0.2)
#X = X*(X<1.2)
```
